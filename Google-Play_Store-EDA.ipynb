## Data Cleaning

### Removing irrelavant columns

```python
#lets remove the "Scrapped Time"
playStore.drop(columns='Scraped Time', inplace=True)
print("The column named 'Scraped Time' got dropped.")
```

### Rectifying the column Data Types

#### 1. Column: Installs
The 'Installs' column is currently of 'object' data type and contains some special characters. We need to change it to numeric data type.

##### Checking for Unique Values

```python
playStore['Installs'].unique()
```

##### Dropping the NaN

```python
#droping the missing values from Installs Column
playStore= playStore.dropna(subset=['Installs'])
new_row_count= playStore.shape[0]
print(f"Number of rows dropped: {curr_row_count-new_row_count}")
curr_row_count= new_row_count
```

##### Removing the '+' sign and ',' using Regex

```python
#the code removes any character that is not "a-zA-Z0-9"
playStore['Installs']= playStore['Installs'].str.replace('[^a-zA-Z0-9]','', regex=True)
```

##### Converting to Integer Data Type 

```python
#each value is converted to integer type
playStore['Installs']= playStore['Installs'].apply(lambda x: int(x))
```

#### 2. Column: Size
The 'Size' column is currently of 'object' data type. We need to change it to numeric data type.

##### Checking for Unique Values

```python
#the Size column has following unique values
playStore['Size'].unique()
```

##### Converting values to Bytes unit
Currently the column contains values in both MB and KB written as '10M' and '3.2k' respectively. We will standardise it to Bytes for easy conversion later on.
Conversion chart:
    1 KB = 1024 Bytes;
    1 MB = 1024 KB

```python
#defining a function to accomplish our task
def convert_to_bytes(size):
    # the function will convert KB to Bytes by multiplying by '1024' and from MB to Bytes by multiplying by '1024*1024'
    if isinstance(size,str):
        #remove any comma in the string
        size= size.replace(',','')
        if 'k' in size:
            return float(size.replace('k','')) * 1024
        elif 'M' in size:
            return float(size.replace('M','')) * 1024 * 1024
    return size

playStore['Size']= playStore['Size'].apply(convert_to_bytes)
```

##### Renaming column and Changing data type

```python
playStore.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
playStore['Size_in_bytes']= pd.to_numeric(playStore['Size_in_bytes'], errors= 'coerce')
```

#### 3. Columns: Released and Last Updated
Changing the data type from Object to Datetime

```python
#in the dataset the Date is in the format of 'Aug 20, 2012' so we will be using "format='%b %d, %Y'"
playStore['Released']= pd.to_datetime(playStore['Released'], format='%b %d, %Y', errors='raise')
playStore['Last Updated']= pd.to_datetime(playStore['Last Updated'], format='%b %d, %Y', errors='raise')
```

### Creating New Columns

#### 1. Adding Column for Size in MB

```python
#dividing the size in bytes by (1024*1024) will give us the required value
playStore['Size_in_MB']= playStore['Size_in_bytes'].apply(lambda x: x/(1024*1024))
print("New column named 'Size_in_MB' was created to store the application size in MB" )
```

```python
#we will use the groupby and transform to create a dataframe of the same same shape and then us these values to fill the Null values
playStore['Size_in_bytes']= playStore['Size_in_bytes'].fillna(playStore.groupby("Category")['Size_in_bytes'].transform('mean')).round(1)
playStore['Size_in_MB']= playStore['Size_in_MB'].fillna(playStore.groupby("Category")['Size_in_MB'].transform('mean')).round(1)
```
